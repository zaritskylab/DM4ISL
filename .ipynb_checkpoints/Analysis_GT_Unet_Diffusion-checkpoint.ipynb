```python
def image_based_metrics_and_measurements(all_GT_images__, all_Unet_images__, all_GAN_images__, all_DDPM_images__, all_DDPMavg_images__):
    rn = 4 ## for  rounding
    print('')
    print('---------------------------')
    print('I2I similarity metrics')
    print('---------------------------')
    GT_Unet_pcc_agg       = pcc(all_GT_images__ , all_Unet_images__   , rn) 
    GT_GAN_pcc_agg        = pcc(all_GT_images__ , all_GAN_images__    , rn) 
    GT_DDPM_pcc_agg       = pcc(all_GT_images__ , all_DDPM_images__   , rn) 
    GT_DDPMavg_pcc_agg    = pcc(all_GT_images__ , all_DDPMavg_images__, rn) 
    GT_Unet_mse_agg       = mse(all_GT_images__ , all_Unet_images__   , rn)  
    GT_GAN_mse_agg        = mse(all_GT_images__ , all_GAN_images__    , rn)  
    GT_DDPM_mse_agg       = mse(all_GT_images__ , all_DDPM_images__   , rn)  
    GT_DDPMavg_mse_agg    = mse(all_GT_images__ , all_DDPMavg_images__, rn)  
    GT_Unet_MSssim_agg    = ms_ssim(all_GT_images__ , all_Unet_images__   , rn)      
    GT_GAN_MSssim_agg     = ms_ssim(all_GT_images__ , all_GAN_images__    , rn)    
    GT_DDPM_MSssim_agg    = ms_ssim(all_GT_images__ , all_DDPM_images__   , rn)
    GT_DDPMavg_MSssim_agg = ms_ssim(all_GT_images__ , all_DDPMavg_images__, rn)
    GT_Unet_js_agg        = jsd(all_GT_images__ , all_Unet_images__   , rn)      
    GT_GAN_js_agg         = jsd(all_GT_images__ , all_GAN_images__    , rn)      
    GT_DDPM_js_agg        = jsd(all_GT_images__ , all_DDPM_images__   , rn)      
    GT_DDPMavg_js_agg     = jsd(all_GT_images__ , all_DDPMavg_images__, rn)      
    GT_Unet_mi_agg        = mi(all_GT_images__ , all_Unet_images__   , rn)          
    GT_GAN_mi_agg         = mi(all_GT_images__ , all_GAN_images__    , rn)           
    GT_DDPM_mi_agg        = mi(all_GT_images__ , all_DDPM_images__   , rn)          
    GT_DDPMavg_mi_agg     = mi(all_GT_images__ , all_DDPMavg_images__, rn)       
    
    print('')
    print( 'pcc    : ' , GT_Unet_pcc_agg , ' ' , GT_GAN_pcc_agg , ' ' , GT_DDPM_pcc_agg , ' ' , GT_DDPMavg_pcc_agg)
    print( 'mse    : ' , GT_Unet_mse_agg , ' ' , GT_GAN_mse_agg , ' ' , GT_DDPM_mse_agg , ' ' , GT_DDPMavg_mse_agg)
    print( 'MSssim : ' , GT_Unet_MSssim_agg , ' ' , GT_GAN_MSssim_agg , ' ' , GT_DDPM_MSssim_agg , ' ' , GT_DDPMavg_MSssim_agg)
    print( 'MI     : ' , GT_Unet_mi_agg , ' ' , GT_GAN_mi_agg , ' ' , GT_DDPM_mi_agg , ' ' , GT_DDPMavg_mi_agg)
    print( 'JS     : ' , GT_Unet_js_agg , ' ' , GT_GAN_js_agg , ' ' , GT_DDPM_js_agg , ' ' , GT_DDPMavg_js_agg)
    
    print('')
    print('---------------------------')
    print('Measurements: texture - on GLCM')
    print('---------------------------')
    all_GT_glcm_correlation = []
    all_Unet_glcm_correlation = []
    all_GAN_glcm_correlation = []
    all_DDPM_glcm_correlation = []
    all_DDPMavg_glcm_correlation = []

    for i in range(all_GT_images__.shape[0]):
        angle=90
        GT_glcm_correlation      = glcm_measurements( all_GT_images__[i], [1], [angle] ) 
        Unet_glcm_correlation    = glcm_measurements( all_Unet_images__[i], [1], [angle])
        GAN_glcm_correlation     = glcm_measurements( all_GAN_images__[i], [1], [angle])
        DDPM_glcm_correlation    = glcm_measurements( all_DDPM_images__[i], [1], [angle])
        DDPMavg_glcm_correlation = glcm_measurements( all_DDPMavg_images__[i], [1], [angle])
        
        all_GT_glcm_correlation.append(GT_glcm_correlation)
        all_Unet_glcm_correlation.append(Unet_glcm_correlation)
        all_GAN_glcm_correlation.append(GAN_glcm_correlation)
        all_DDPM_glcm_correlation.append(DDPM_glcm_correlation)
        all_DDPMavg_glcm_correlation.append(DDPMavg_glcm_correlation)
    all_GT_glcm_correlation      = np.array(all_GT_glcm_correlation)
    all_Unet_glcm_correlation    = np.array(all_Unet_glcm_correlation)
    all_GAN_glcm_correlation     = np.array(all_GAN_glcm_correlation)
    all_DDPM_glcm_correlation    = np.array(all_DDPM_glcm_correlation)
    all_DDPMavg_glcm_correlation = np.array(all_DDPMavg_glcm_correlation)

    GT_Unet_glcm_correlation   = np.round(stats.pearsonr( all_GT_glcm_correlation   , all_Unet_glcm_correlation  )[0], rn)                                  
    GT_GAN_glcm_correlation   = np.round(stats.pearsonr( all_GT_glcm_correlation   , all_GAN_glcm_correlation  )[0], rn)   
    GT_DDPM_glcm_correlation   = np.round(stats.pearsonr( all_GT_glcm_correlation   , all_DDPM_glcm_correlation  )[0], rn) 
    GT_DDPMavg_glcm_correlation   = np.round(stats.pearsonr( all_GT_glcm_correlation   , all_DDPMavg_glcm_correlation  )[0], rn)                                 
    
    print('')  
    print( 'glcm_matrix_correlation_pcc   : ' , GT_Unet_glcm_correlation   , GT_GAN_glcm_correlation   , GT_DDPM_glcm_correlation   , GT_DDPMavg_glcm_correlation    )
  

    ### HOG
    all_GT_hog_features, all_Unet_hog_features, all_GAN_hog_features, all_DDPM_hog_features, all_DDPMavg_hog_features = [],[],[],[],[]
    for i in range(all_GT_images__.shape[0]):  
        GT_hog_features, Unet_hog_features, GAN_hog_features, DDPM_hog_features, DDPMavg_hog_features = [],[],[],[],[]
        for slice in range (all_GT_images__[i].shape[0]):
            GT_hog_features.append(   hog(all_GT_images__[i][slice]  , orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2)) ) # (16, 1764)
            Unet_hog_features.append( hog(all_Unet_images__[i][slice], orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2)) ) 
            GAN_hog_features.append( hog(all_GAN_images__[i][slice], orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2)) ) 
            DDPM_hog_features.append( hog(all_DDPM_images__[i][slice], orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2)) ) 
            DDPMavg_hog_features.append( hog(all_DDPMavg_images__[i][slice], orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2)) ) 

        GT_hog_features, Unet_hog_features, GAN_hog_features, DDPM_hog_features, DDPMavg_hog_features = np.array(GT_hog_features), np.array(Unet_hog_features), np.array(GAN_hog_features), np.array(DDPM_hog_features), np.array(DDPMavg_hog_features)
        all_GT_hog_features.append( GT_hog_features ) , all_Unet_hog_features.append( Unet_hog_features ) , all_GAN_hog_features.append( GAN_hog_features ) , all_DDPM_hog_features.append( DDPM_hog_features ) , all_DDPMavg_hog_features.append( DDPMavg_hog_features )
    all_GT_hog_features, all_Unet_hog_features, all_GAN_hog_features , all_DDPM_hog_features, all_DDPMavg_hog_features = np.array(all_GT_hog_features), np.array(all_Unet_hog_features), np.array(all_GAN_hog_features), np.array(all_DDPM_hog_features), np.array(all_DDPMavg_hog_features)
    print('')
    print( 'HOG: ' , np.round( np.mean(np.abs(all_GT_hog_features - all_Unet_hog_features)),rn) , np.round( np.mean(np.abs(all_GT_hog_features - all_GAN_hog_features)),rn) , np.round( np.mean(np.abs(all_GT_hog_features - all_DDPM_hog_features)),rn), np.round( np.mean(np.abs(all_GT_hog_features - all_DDPMavg_hog_features)),rn) )
    
    ### using CLS
    print('')          
    GT_Unet_CLSperc    = lpips(main_path, torch.tensor(all_GT_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda') , torch.tensor(all_Unet_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda'))
    GT_GAN_CLSperc     = lpips(main_path, torch.tensor(all_GT_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda') , torch.tensor(all_GAN_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda'))
    GT_DDPM_CLSperc    = lpips(main_path, torch.tensor(all_GT_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda') , torch.tensor(all_DDPM_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda'))
    GT_DDPMavg_CLSperc = lpips(main_path, torch.tensor(all_GT_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda') , torch.tensor(all_DDPMavg_images__).permute(0,2,3,1).unsqueeze(1).to(torch.float32).to('cuda'))

    print('CLS LPIPS ' , np.round(GT_Unet_CLSperc.mean(), rn) , np.round(GT_GAN_CLSperc.mean(), rn) , np.round(GT_DDPM_CLSperc.mean(), rn) , np.round(GT_DDPMavg_CLSperc.mean(), rn)   ) ## 5 layers: edges, textures, pattens, objects, pooled

    
    ###############################################################################
    ######## segmentation ######################################################### 
    ###############################################################################
    print('')
    print('---------------------------')
    print('segmentation')
    print('---------------------------')

    # all_GT_images__, all_Unet_images__, all_GAN_images__, all_DDPM_images__, all_DDPMavg_images__ = GT_images, Unet_images, GAN_images, DDPM_images, DDPMavg_images

    ## segmentation per patch3D
    GT_seg_morph_slices, Unet_seg_morph_slices, GAN_seg_morph_slices, DDPM_seg_morph_slices, DDPMavg_seg_morph_slices = np.zeros_like(all_GT_images__), np.zeros_like(all_GT_images__), np.zeros_like(all_GT_images__), np.zeros_like(all_GT_images__), np.zeros_like(all_GT_images__)
    for i in range(all_GT_images__.shape[0]):    
        GT_th, GT_seg_morph_slices_           = segmentation_pipeline(all_GT_images__[i]      , filter_type=seg_params.filter_type, k1=seg_params.k1, k2=seg_params.k2, k3=seg_params.k3, filter_kernel=seg_params.filter_kernel, sigma=seg_params.sigma, organelle_th=seg_params.organelle_th, do_erode_dilate=seg_params.do_erode_dilate, do_remove_small_objects=seg_params.do_remove_small_objects, do_fill_holes=seg_params.do_fill_holes, do_fill_holes_boarders=seg_params.do_fill_holes_boarders)  
        Unet_th, Unet_seg_morph_slices_       = segmentation_pipeline(all_Unet_images__[i]    , filter_type=seg_params.filter_type, k1=seg_params.k1, k2=seg_params.k2, k3=seg_params.k3, filter_kernel=seg_params.filter_kernel, sigma=seg_params.sigma, organelle_th=seg_params.organelle_th, do_erode_dilate=seg_params.do_erode_dilate, do_remove_small_objects=seg_params.do_remove_small_objects, do_fill_holes=seg_params.do_fill_holes, do_fill_holes_boarders=seg_params.do_fill_holes_boarders)  
        GAN_th, GAN_seg_morph_slices_         = segmentation_pipeline(all_GAN_images__[i]     , filter_type=seg_params.filter_type, k1=seg_params.k1, k2=seg_params.k2, k3=seg_params.k3, filter_kernel=seg_params.filter_kernel, sigma=seg_params.sigma, organelle_th=seg_params.organelle_th, do_erode_dilate=seg_params.do_erode_dilate, do_remove_small_objects=seg_params.do_remove_small_objects, do_fill_holes=seg_params.do_fill_holes, do_fill_holes_boarders=seg_params.do_fill_holes_boarders)  
        DDPM_th, DDPM_seg_morph_slices_       = segmentation_pipeline(all_DDPM_images__[i]    , filter_type=seg_params.filter_type, k1=seg_params.k1, k2=seg_params.k2, k3=seg_params.k3, filter_kernel=seg_params.filter_kernel, sigma=seg_params.sigma, organelle_th=seg_params.organelle_th, do_erode_dilate=seg_params.do_erode_dilate, do_remove_small_objects=seg_params.do_remove_small_objects, do_fill_holes=seg_params.do_fill_holes, do_fill_holes_boarders=seg_params.do_fill_holes_boarders)
        DDPMavg_th, DDPMavg_seg_morph_slices_ = segmentation_pipeline(all_DDPMavg_images__[i] , filter_type=seg_params.filter_type, k1=seg_params.k1, k2=seg_params.k2, k3=seg_params.k3, filter_kernel=seg_params.filter_kernel, sigma=seg_params.sigma, organelle_th=seg_params.organelle_th, do_erode_dilate=seg_params.do_erode_dilate, do_remove_small_objects=seg_params.do_remove_small_objects, do_fill_holes=seg_params.do_fill_holes, do_fill_holes_boarders=seg_params.do_fill_holes_boarders)
        GT_seg_morph_slices[i] = GT_seg_morph_slices_
        Unet_seg_morph_slices[i] = Unet_seg_morph_slices_
        GAN_seg_morph_slices[i] = GAN_seg_morph_slices_
        DDPM_seg_morph_slices[i] = DDPM_seg_morph_slices_
        DDPMavg_seg_morph_slices[i] = DDPMavg_seg_morph_slices_

    ### segmentation masks 3D 
    GT_seg_morph_slices = (GT_seg_morph_slices//255).astype('uint8') # 0 or 1
    Unet_seg_morph_slices = (Unet_seg_morph_slices//255).astype('uint8')
    GAN_seg_morph_slices = (GAN_seg_morph_slices//255).astype('uint8')
    DDPM_seg_morph_slices = (DDPM_seg_morph_slices//255).astype('uint8')
    DDPMavg_seg_morph_slices = (DDPMavg_seg_morph_slices//255).astype('uint8')

    TP, FN, FP, GT_Unet_iou, GT_Unet_dice, GT_Unet_precision, GT_Unet_recall, GT_Unet_F1 = seg_metrics( GT_seg_morph_slices , Unet_seg_morph_slices ) 
    TP, FN, FP, GT_GAN_iou, GT_GAN_dice, GT_GAN_precision, GT_GAN_recall, GT_GAN_F1      = seg_metrics( GT_seg_morph_slices , GAN_seg_morph_slices ) 
    TP, FN, FP, GT_DDPM_iou, GT_DDPM_dice, GT_DDPM_precision, GT_DDPM_recall, GT_DDPM_F1 = seg_metrics( GT_seg_morph_slices , DDPM_seg_morph_slices ) 
    TP, FN, FP, GT_DDPMavg_iou, GT_DDPMavg_dice, GT_DDPMavg_precision, GT_DDPMavg_recall, GT_DDPMavg_F1 = seg_metrics( GT_seg_morph_slices , DDPMavg_seg_morph_slices ) 

    print('')
    print( 'IoU       : ' , np.round(GT_Unet_iou, rn)      , ' ' , np.round(GT_GAN_iou, rn)      , ' ' , np.round(GT_DDPM_iou, rn)      , ' ' , np.round(GT_DDPMavg_iou, rn)      )
    print( 'Dice      : ' , np.round(GT_Unet_dice,rn)      , ' ' , np.round(GT_GAN_dice,rn)      , ' ' , np.round(GT_DDPM_dice,rn)      , ' ' , np.round(GT_DDPMavg_dice,rn)      )
    print( 'Precision : ' , np.round(GT_Unet_precision,rn) , ' ' , np.round(GT_GAN_precision,rn) , ' ' , np.round(GT_DDPM_precision,rn) , ' ' , np.round(GT_DDPMavg_precision,rn) )
    print( 'Recall    : ' , np.round(GT_Unet_recall,rn)    , ' ' , np.round(GT_GAN_recall,rn)    , ' ' , np.round(GT_DDPM_recall,rn)    , ' ' , np.round(GT_DDPMavg_recall,rn)    )

    #### downstream measurements
    GT_Unet_volume_diff,GT_Unet_cx_diff,GT_Unet_cy_diff,GT_Unet_cz_diff,GT_Unet_cr_diff = 0,0,0,0,0
    GT_GAN_volume_diff,GT_GAN_cx_diff,GT_GAN_cy_diff,GT_GAN_cz_diff,GT_GAN_cr_diff      = 0,0,0,0,0
    GT_DDPM_volume_diff,GT_DDPM_cx_diff,GT_DDPM_cy_diff,GT_DDPM_cz_diff,GT_DDPM_cr_diff = 0,0,0,0,0
    GT_DDPMavg_volume_diff,GT_DDPMavg_cx_diff,GT_DDPMavg_cy_diff,GT_DDPMavg_cz_diff,GT_DDPMavg_cr_diff = 0,0,0,0,0
    GT_Unet_hd, GT_GAN_hd ,GT_DDPM_hd, GT_DDPMavg_hd = 0,0,0,0
    for i in range(all_GT_images__.shape[0]):
        GT_binary3D_volume, GT_binary3D_cz, GT_binary3D_cx, GT_binary3D_cy, GT_binary3D_cr = get_markers_measurements( GT_seg_morph_slices[i] )
        Unet_binary3D_volume, Unet_binary3D_cz, Unet_binary3D_cx, Unet_binary3D_cy, Unet_binary3D_cr = get_markers_measurements( Unet_seg_morph_slices[i] )
        GAN_binary3D_volume, GAN_binary3D_cz, GAN_binary3D_cx, GAN_binary3D_cy, GAN_binary3D_cr = get_markers_measurements( GAN_seg_morph_slices[i] )
        DDPM_binary3D_volume, DDPM_binary3D_cz, DDPM_binary3D_cx, DDPM_binary3D_cy, DDPM_binary3D_cr = get_markers_measurements( DDPM_seg_morph_slices[i] )
        DDPMavg_binary3D_volume, DDPMavg_binary3D_cz, DDPMavg_binary3D_cx, DDPMavg_binary3D_cy, DDPMavg_binary3D_cr = get_markers_measurements( DDPMavg_seg_morph_slices[i] )

        GT_Unet_volume_diff = np.append(GT_Unet_volume_diff , np.abs(1 - GT_binary3D_volume[0] / Unet_binary3D_volume[0]) )
        GT_Unet_cx_diff     = np.append(GT_Unet_cx_diff     , np.abs(1 - GT_binary3D_cx[0] / Unet_binary3D_cx[0]) )
        GT_Unet_cy_diff     = np.append(GT_Unet_cy_diff     , np.abs(1 - GT_binary3D_cy[0] / Unet_binary3D_cy[0]) )
        GT_Unet_cz_diff     = np.append(GT_Unet_cz_diff     , np.abs(1 - GT_binary3D_cz[0] / Unet_binary3D_cz[0]) )
        GT_Unet_cr_diff     = np.append(GT_Unet_cr_diff     , np.abs(1 - GT_binary3D_cr[0] / Unet_binary3D_cr[0]) )
        GT_GAN_volume_diff  = np.append(GT_GAN_volume_diff , np.abs(1 - GT_binary3D_volume[0] / GAN_binary3D_volume[0]) )
        GT_GAN_cx_diff      = np.append(GT_GAN_cx_diff     , np.abs(1 - GT_binary3D_cx[0]     / GAN_binary3D_cx[0]) )
        GT_GAN_cy_diff      = np.append(GT_GAN_cy_diff     , np.abs(1 - GT_binary3D_cy[0]     / GAN_binary3D_cy[0]) )
        GT_GAN_cz_diff      = np.append(GT_GAN_cz_diff     , np.abs(1 - GT_binary3D_cz[0]     / GAN_binary3D_cz[0]) )
        GT_GAN_cr_diff      = np.append(GT_GAN_cr_diff     , np.abs(1 - GT_binary3D_cr[0]     / GAN_binary3D_cr[0]) )
        GT_DDPM_volume_diff = np.append(GT_DDPM_volume_diff , np.abs(1 - GT_binary3D_volume[0] / DDPM_binary3D_volume[0]) )
        GT_DDPM_cx_diff     = np.append(GT_DDPM_cx_diff     , np.abs(1 - GT_binary3D_cx[0] / DDPM_binary3D_cx[0]) )
        GT_DDPM_cy_diff     = np.append(GT_DDPM_cy_diff     , np.abs(1 - GT_binary3D_cy[0] / DDPM_binary3D_cy[0]) )
        GT_DDPM_cz_diff     = np.append(GT_DDPM_cz_diff     , np.abs(1 - GT_binary3D_cz[0] / DDPM_binary3D_cz[0]) )
        GT_DDPM_cr_diff     = np.append(GT_DDPM_cr_diff     , np.abs(1 - GT_binary3D_cr[0] / DDPM_binary3D_cr[0]) )
        GT_DDPMavg_volume_diff = np.append(GT_DDPMavg_volume_diff , np.abs(1 - GT_binary3D_volume[0] / DDPMavg_binary3D_volume[0]) )
        GT_DDPMavg_cx_diff     = np.append(GT_DDPMavg_cx_diff     , np.abs(1 - GT_binary3D_cx[0]     / DDPMavg_binary3D_cx[0]) )
        GT_DDPMavg_cy_diff     = np.append(GT_DDPMavg_cy_diff     , np.abs(1 - GT_binary3D_cy[0]     / DDPMavg_binary3D_cy[0]) )
        GT_DDPMavg_cz_diff     = np.append(GT_DDPMavg_cz_diff     , np.abs(1 - GT_binary3D_cz[0]     / DDPMavg_binary3D_cz[0]) )
        GT_DDPMavg_cr_diff     = np.append(GT_DDPMavg_cr_diff     , np.abs(1 - GT_binary3D_cr[0]     / DDPMavg_binary3D_cr[0]) )

        GT_contours_coor, GT_contours_imgs = make_contours_from_segmentation( GT_seg_morph_slices[i] )
        Unet_contours_coor, Unet_contours_imgs = make_contours_from_segmentation( Unet_seg_morph_slices[i] )
        GAN_contours_coor, GAN_contours_imgs = make_contours_from_segmentation( GAN_seg_morph_slices[i] )
        DDPM_contours_coor, DDPM_contours_imgs = make_contours_from_segmentation( DDPM_seg_morph_slices[i] )
        DDPMavg_contours_coor, DDPMavg_contours_imgs = make_contours_from_segmentation( DDPMavg_seg_morph_slices[i] )
        GT_Unet_hd = np.append( GT_Unet_hd , hausdorff_distance(volumetric2sequence( GT_contours_imgs ) , volumetric2sequence( Unet_contours_imgs )) )   
        GT_GAN_hd = np.append( GT_GAN_hd , hausdorff_distance(volumetric2sequence( GT_contours_imgs ) , volumetric2sequence( GAN_contours_imgs )) )   
        GT_DDPM_hd = np.append( GT_DDPM_hd , hausdorff_distance(volumetric2sequence( GT_contours_imgs ) , volumetric2sequence( DDPM_contours_imgs )) )   
        GT_DDPMavg_hd = np.append( GT_DDPMavg_hd , hausdorff_distance(volumetric2sequence( GT_contours_imgs ) , volumetric2sequence( DDPMavg_contours_imgs )) )  

    print('HD        : ' , np.round(GT_Unet_hd[np.isfinite(GT_Unet_hd)][1:].mean(),rn) , ' ' , np.round(GT_GAN_hd[np.isfinite(GT_GAN_hd)][1:].mean(),rn) , ' ' , np.round(GT_DDPM_hd[np.isfinite(GT_DDPM_hd)][1:].mean(),rn) , ' ' , np.round(GT_DDPMavg_hd[np.isfinite(GT_DDPMavg_hd)][1:].mean(),rn)   )

    GT_Unet_volume_diff,GT_Unet_cx_diff,GT_Unet_cy_diff,GT_Unet_cz_diff,GT_Unet_cr_diff = GT_Unet_volume_diff[1:].mean(),GT_Unet_cx_diff[1:].mean(),GT_Unet_cy_diff[1:].mean(),GT_Unet_cz_diff[1:].mean(),GT_Unet_cr_diff[1:].mean()
    GT_GAN_volume_diff,GT_GAN_cx_diff,GT_GAN_cy_diff,GT_GAN_cz_diff,GT_GAN_cr_diff      = GT_GAN_volume_diff[1:].mean(),GT_GAN_cx_diff[1:].mean(),GT_GAN_cy_diff[1:].mean(),GT_GAN_cz_diff[1:].mean(),GT_GAN_cr_diff[1:].mean()
    GT_DDPM_volume_diff,GT_DDPM_cx_diff,GT_DDPM_cy_diff,GT_DDPM_cz_diff,GT_DDPM_cr_diff = GT_DDPM_volume_diff[1:].mean(),GT_DDPM_cx_diff[1:].mean(),GT_DDPM_cy_diff[1:].mean(),GT_DDPM_cz_diff[1:].mean(),GT_DDPM_cr_diff[1:].mean()
    GT_DDPMavg_volume_diff,GT_DDPMavg_cx_diff,GT_DDPMavg_cy_diff,GT_DDPMavg_cz_diff,GT_DDPMavg_cr_diff = GT_DDPMavg_volume_diff[1:].mean(),GT_DDPMavg_cx_diff[1:].mean(),GT_DDPMavg_cy_diff[1:].mean(),GT_DDPMavg_cz_diff[1:].mean(),GT_DDPMavg_cr_diff[1:].mean()
    print('GT Unet: ' , 'volume ' , np.round(GT_Unet_volume_diff,rn) , 'cr ' , np.round(GT_Unet_cr_diff,rn) , 'cx ' , np.round(GT_Unet_cx_diff,rn) , 'cy ' , np.round(GT_Unet_cy_diff,rn) , 'cz ' , np.round(GT_Unet_cz_diff,rn)  )
    print('GT GAN: ' , 'volume ' , np.round(GT_Unet_volume_diff,rn) , 'cr ' , np.round(GT_GAN_cr_diff,rn) , 'cx ' , np.round(GT_GAN_cx_diff,rn) , 'cy ' , np.round(GT_GAN_cy_diff,rn) , 'cz ' , np.round(GT_GAN_cz_diff,rn)  )
    print('GT DDPM: ' , 'volume ' , np.round(GT_DDPM_volume_diff,rn) , 'cr ' , np.round(GT_DDPM_cr_diff,rn) , 'cx ' , np.round(GT_DDPM_cx_diff,rn) , 'cy ' , np.round(GT_DDPM_cy_diff,rn) , 'cz ' , np.round(GT_DDPM_cz_diff,rn)  )
    print('GT DDPMavg: ' , 'volume ' , np.round(GT_DDPMavg_volume_diff,rn) , 'cr ' , np.round(GT_DDPMavg_cr_diff,rn) , 'cx ' , np.round(GT_DDPMavg_cx_diff,rn) , 'cy ' , np.round(GT_DDPMavg_cy_diff,rn) , 'cz ' , np.round(GT_DDPMavg_cz_diff,rn)  )

```
